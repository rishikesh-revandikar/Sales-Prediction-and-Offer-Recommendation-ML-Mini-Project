```python
import pandas as pd
import numpy as np
```

# DATA CLEANING

```python
df.isnull().sum()
```

```python
df['Item_Weight'].mean()
```

```python
df['Item_Weight'].fillna(df['Item_Weight'].mean(),inplace=True)
```

```python
df['Outlet_Size'].mode()[0]
```

```python
df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Item_Type_Combined'] = df['Item_Identifier'].apply(lambda x: x[0:2])
#Rename them to more intuitive categories:
df['Item_Type_Combined'] = df['Item_Type_Combined'].map({'FD':'Food',
                                                             'NC':'Non-Consumable',
                                                             'DR':'Drinks'})
df['Item_Type_Combined'].value_counts()
```

```python
df
```

```python
print('Original Categories:')
print(df['Item_Fat_Content'].value_counts())

print('\nModified Categories:')
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({'LF':'Low Fat',
                                                             'reg':'Regular',
                                                             'low fat':'Low Fat'})
print(df['Item_Fat_Content'].value_counts())
```

```python
# df['Years_Established'] = df['Outlet_Establishment_Year'].apply(lambda x: 2023 - x) 
```

```python
df
```

```python
ui = df['Outlet_Identifier'].unique()
```

```python
year = df['Outlet_Establishment_Year'].unique()
```

```python
print('oi : ',ui)
print('year : ',year)
```

```python
df.columns
```

```python
df['Item_Visibility'].value_counts(0)
```

```python
df['Item_Visibility']=df['Item_Visibility'].replace(0,df['Item_Visibility'].mean())
```

```python
df['Item_Visibility'].value_counts(0)
```

```python
colums_to_drop = ['Item_Visibility','Item_Identifier','Outlet_Identifier','Outlet_Establishment_Year']
```

```python
df.drop(colums_to_drop,axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

s1 = StandardScaler()
df[['Item_Weight']] = s1.fit_transform(df[['Item_Weight']])


s2 = StandardScaler()
df[['Item_MRP']] = s2.fit_transform(df[['Item_MRP']])
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python

l1 =  LabelEncoder()
l2 =  LabelEncoder()
l3 =  LabelEncoder()
l4 =  LabelEncoder()
l5 =  LabelEncoder()
l6 =  LabelEncoder()
df['Item_Fat_Content'] = l1.fit_transform(df['Item_Fat_Content'])
df['Item_Type'] = l2.fit_transform(df['Item_Type'])
df['Outlet_Size'] = l3.fit_transform(df['Outlet_Size'])
df['Outlet_Location_Type'] = l4.fit_transform(df['Outlet_Location_Type'])
df['Outlet_Type'] = l5.fit_transform(df['Outlet_Type'])
df['Item_Type_Combined'] = l6.fit_transform(df['Item_Type_Combined'])
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['Item_Fat_Content','Outlet_Size','Outlet_Location_Type','Outlet_Type','Item_Type_Combined'])
```

```python
df
```

```python
# Define a list of columns you want to convert to 0s and 1s
columns_to_convert = ['Item_Fat_Content_0','Item_Fat_Content_1','Outlet_Size_0','Outlet_Size_1','Outlet_Size_2',
                      'Outlet_Location_Type_0','Outlet_Location_Type_1','Outlet_Location_Type_2','Outlet_Type_0',
                      'Outlet_Type_1','Outlet_Type_2','Outlet_Type_3','Item_Type_Combined_0','Item_Type_Combined_1',
                     'Item_Type_Combined_2']

# Apply the conversion using apply and lambda
df[columns_to_convert] = df[columns_to_convert].apply(lambda x: x.astype(int))
```

# DATA SPLITTING

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Item_Outlet_Sales'],axis=1)
Y = df['Item_Outlet_Sales']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state=101,test_size=0.2)
```

# MODEL BUILDING

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
y_pred = LR.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF = RandomForestRegressor()
```

```python
RF.fit(x_train,y_train)
```

```python
y_pred_rf=RF.predict(x_test)
```

```python
y_pred_rf
```

```python
print(r2_score(y_test,y_pred_rf))
print(mean_absolute_error(y_test,y_pred_rf))
print(np.sqrt(mean_squared_error(y_test,y_pred_rf)))
```
